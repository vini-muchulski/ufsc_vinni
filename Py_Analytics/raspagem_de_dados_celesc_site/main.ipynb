```python
def formatar_dado(valor):
    
    # Remover espaços em branco no início e no fim da linha
    line = valor.strip()
    
    # Verificar se a linha não está vazia
    if line:
        # Encontrar o último espaço na linha para separar o tipo de receita e o valor
        last_space_index = line.rfind(' ')
        tipo_receita = line[:last_space_index].strip()
        valor_receita = line[last_space_index:].strip().replace('.', '').replace(',', '.')

        return tipo_receita, valor_receita
    else:
        print(f"erro ao formatar {valor}")
        
```

```python

output_file= "valores_receitas.csv"

csvfile = open(output_file, 'w', newline='', encoding='utf-8')
csvwriter = csv.writer(csvfile)
    
# Escrever o cabeçalho
csvwriter.writerow(['Tipo de Receita', 'Valor da Receita'])
```

```python

for tabela in tabelas:
    linhas = tabela.find_elements(By.TAG_NAME, 'tr')
    for linha in linhas:
        data = []
        cols = linha.find_elements(By.TAG_NAME, 'td')
        
        tipo_receita = cols[0].text.strip()
        valor_receita = cols[1].text.strip().replace('.', '').replace(',', '.')

        #em economia numeros com () sao negativos
        if valor_receita.startswith('(') and valor_receita.endswith(')'):
                    valor_receita = '-' + valor_receita[1:-1]
            
        valor_receita = float(valor_receita)
        
        data.append([tipo_receita, valor_receita])
        print(data)
        
        
        csvwriter.writerows(data)

csvfile.close()
```
