# Tipo de concorrencia

```python
df_saude = df[df['Unidade Gestora'].str.contains('Fundo Municipal de Saude', case=False, na=False)]
```

```python
df_saude.shape
```

```python
df_saude.head()
```

```python
# Contar a distribuição dos tipos de concorrência
distribuicao_concorrencia = df_saude['Tipo de Concorrência'].value_counts()

print(distribuicao_concorrencia)

```

```python
distribuicao_concorrencia.plot(kind='bar')
plt.title('Distribuição dos Tipos de Concorrência na Área da Saúde')
plt.xlabel('Tipo de Concorrência')
plt.ylabel('Número de Licitações')
plt.show()
```

```python
#  Análise de Custos Estimados vs. Homologados
df_saude['Valor - Estimado'] = pd.to_numeric(df_saude['Valor - Estimado'], errors='coerce')
df_saude['Valor - Homologado'] = pd.to_numeric(df_saude['Valor - Homologado'], errors='coerce')

```

```python
# Criar um gráfico para comparar os valores
df_saude[['Valor - Estimado', 'Valor - Homologado']].plot(kind='bar')
plt.title('Comparação entre Valor Estimado e Valor Homologado - Saúde')
plt.ylabel('Valor (R$)')
plt.show()
```

# Razao entre valores de licitacao da saude vs total

```python


# Filtrar as licitações relacionadas à saúde
df_saude = df[df['Unidade Gestora'].str.contains('Fundo Municipal de Saude', case=False, na=False)]


total_valor_estimado_saude = df_saude['Valor - Estimado'].sum()


df_outros = df
total_valor_estimado_outros = df_outros['Valor - Estimado'].sum()

razao_saude_vs_outros = total_valor_estimado_saude / total_valor_estimado_outros


```

```python

print(f"Total valor estimado - Saúde: R$ {total_valor_estimado_saude:,.2f}")
print(f"Total valor estimado - Outros: R$ {total_valor_estimado_outros:,.2f}")
print(f"Razão (Saúde / Outros): {razao_saude_vs_outros:.2f}")
```
