```python
bens["Valor Cont치bil"] = pd.to_numeric(bens["Valor Cont치bil"], errors='coerce')
```

```python
tipos_desejados = ["Bens Im칩veis", "Bens M칩veis", "Bens de Dom칤nio P칰blico"]

# Filtra o DataFrame 'bens' mantendo apenas os tipos desejados
bens_filtrados = bens[bens["Tipo"].isin(tipos_desejados)]

bens_filtrados
```

```python
#MAIORES FORNECEDORES
valor_bens_por_fornecedor = bens_filtrados.groupby("Fornecedor")["Valor Cont치bil"].sum()

# Ordenando em ordem decrescente
valor_bens_por_fornecedor = valor_bens_por_fornecedor.sort_values(ascending=False)
valor_bens_por_fornecedor
```
